```python
price_by_airconditioning = house.groupby('airconditioning')['price'].mean()
print(price_by_airconditioning)
```

```python
mean_price = house['price'].mean()
median_price = house['price'].median()
price_75th_percentile = house['price'].quantile(0.75)
print("Mean price:", mean_price)
print("Median price:", median_price)
print("75th percentile price:", price_75th_percentile)
```

```python
moderate_furnished_houses = house[(house['price'] >= median_price - 100000) &
                                (house['price'] <= median_price + 100000) &
                                (house['bedrooms'] >= 3) &
                                (house['bathrooms'] >= 2) &
                                (house['furnishingstatus'] == 1)]
print(moderate_furnished_houses)
```

```python
filtered_houses = house[(house['furnishingstatus'] == 'furnished') &
                        (house['airconditioning'] == 'yes') &
                        (house['mainroad'] == 'yes') &
                        (house['guestroom']=='yes') &
                        (house['basement']=='yes') ]\
                        .sort_values('price')

filtered_houses
```

In this analysis we can easily find the house having all the facilities like furnished AC guestroom etc corresponding to price.
